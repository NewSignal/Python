# POJ1651题目
题目大意是给出N个数，每次从中抽出一个数(第一和最后一个不能抽)，该次的得分即为抽出的数与相邻两个数的乘积。一直这样将每次的得分累加直到只剩下首尾两个数为止，问最小得分。

```python
from numpy import *
```

```python
#N个数
N=10
a=[]
#N个数具体的数值存放在n中
n=[]
#初始化一个N*N的矩阵，并填充0
for i in range(N):
    tem_a=[]
    for j in range(N):
        tem_a.append(0)
    a.append(tem_a)
```

```python
#随机在1到100生成个数存放在n中
for i in range(N):
    n.append(random.randint(1,100))
print(n)
for i in range(N-2):
    a[i][i+2]=n[i]*n[i+1]*n[i+2]
a
```

```python
#len是步长
#下面的代码通过截图和说明更好理解
for len in range(3,N):
    print("len:",len)
    for i in range(N-len):
        print(i)
        j=i+len
        for k in range(i+1,j):
            print(i,j)
            print(i,k,"+",k,j,"+",i,"*",k,"*",j)
            if a[i][j]==0:
                a[i][j]=a[i][k]+a[k][j]+n[i]*n[k]*n[j]
            else:
                a[i][j]=min(a[i][j],a[i][k]+a[k][j]+n[i]*n[k]*n[j])
print(a[0][N-1])
```
