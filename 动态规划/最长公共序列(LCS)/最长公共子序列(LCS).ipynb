# 问题描述
若给定序列X={x1,x2,…,xm}，则另一序列Z={z1,z2,…,zk}是X的子序列是指存在一个严格递增下标序列{i1,i2,…,ik}使得对于所有j=1,2,…,k有：zj=xij(j是i的下标)。给定2个序列X和Y，当另一序列Z既是X的子序列又是Y的子序列时，称Z是序列X
和Y的公共子序列。

## POJ1458
c[i][j]=$\begin{equation}
\left\{
\begin{aligned}
0 \qquad i=0 or j=0 \\
c[i-1,j-1]+1 \qquad i,j>0 and x_{i}=y_{j} \\ 
max(c[i,j-1],c[i-1,j]) \qquad  i,j＞0 and x_{i}\neq y_{j} \\
\end{aligned}
\right.
\end{equation} $

```python
from numpy import *
```

```python
def issame(a,b):
    if a==b:
        return 1
    else:
        return 0
```

```python
#初始化2个数组s1,s2
s1=[random.randint(10) for i in range(random.randint(15,20))]
s2=[random.randint(10) for i in range(random.randint(10,15))]
```

```python
for i in s1:
    print(i)
print(s2)
```

```python
dp=[([0]*20) for i in range(20)]
dp
```

```python
len1=len(s1)
len2=len(s2)
for i in range(1,len1+1):
    for j in range(1,len2+1):
        dp[i][j]=max(dp[i-1][j-1]+issame(s1[i-1],s2[j-1]),dp[i-1][j],dp[i][j-1])
dp[len1][len2]
```
